```
def my(data1, data2, data3, connect):
  conn = sqlite3.connect(f'{connect}')
  df1 = pd.read_csv(f'{data1}')
  df2 = pd.read_csv(f'{data2}',sep=';', header=None, names=['Age', 'City', 'Gender', 'FirstName', 'Email', 'LastName', 'UserName', 'Country'])
  df3 = pd.read_csv(f'{data3}',sep='\t', skiprows=1, names=['Gender', 'FirstName', 'Email', 'Age', 'City', 'Country', 'LastName', 'UserName'])

  new = df2.FirstName.str.split(expand=True)
  df2.FirstName, df2.LastName, df2.UserName = new[0], new[1] ,new[0]
  new2 = df3.FirstName.str.split(expand=True)
  df3.Gender, df3.FirstName, df3.LastName, df3.UserName, df3.Email, df3.Age, df3.City = df3.Gender.str.replace('\w+_',''), new2[0].str.replace('\w+_',''), new2[1] ,new2[0].str.replace('\w+_',''), df3.Email.str.replace('\w+g_',''), df3.Age.str.replace('\w+_',''), df3.City.str.replace('\w+g_','')

  df = pd.concat([df1,df2,df3])

  gender = {'0': 'Female', '1': 'Male', 'F' : 'Female', 'M': 'Male'}
  df.Gender, df.FirstName, df.LastName, df.UserName, df.Email, df.Age, df.City, df.Country = df.Gender.replace(gender), df.FirstName.str.replace('\W', ''), df.LastName.str.replace('\W', ''), df.UserName.str.replace('\W', ''), df.Email.str.lower(), df.Age.str.replace('\D', ''), df.City.str.replace('_', '-'), 'USA'
  df.FirstName, df.LastName, df.UserName, df.City = df.FirstName.str.title(), df.LastName.str.title(), df.UserName.str.lower(), df.City.str.title()
  
  return df.to_sql('customers', conn, index=False)

my('only_wood_customer_us_1.csv', 'only_wood_customer_us_2.csv', 'only_wood_customer_us_3.csv','plastic_free_boutique.sql')
```

```

def gender(data):
    gender = {'0': 'Female', '1': 'Male', 'F' : 'Female', 'M': 'Male'}
    data.Gender = data.Gender.replace(gender)
    return data.Gender

def names(data,column):
    data[f'{column}'] = data[f'{column}'].str.replace('\W', '')
    data[f'{column}'] = data[f'{column}'].str.title()
    return data[f'{column}']

def age(data,column):
    data[f'{column}'] = data[f'{column}'].str.replace('\D', '')
    return data[f'{column}']

def city(data,column):
    data[f'{column}'] = data[f'{column}'].str.replace('_', '-')
    data[f'{column}'] = data[f'{column}'].str.title()
    return data[f'{column}']

def country(data, column):
    city = {'12':'United State Of America', 'U.S.A.':'United State Of America', 'US':'United State Of America','USA':'United State Of America','usa': 'United State Of America', 'u.s.': 'United State Of America', 'U.S.': 'United State Of America','United state of america': 'United State Of America', 'us': 'United State Of America', 'United-State-Of-America': 'United State Of America','U.s.a.': 'United State Of America', 'U.s.': 'United State Of America', 'UNITED STATE OF AMERICA': 'United State Of America', 'u.s.a.': 'United State Of America','united state of america': 'United State Of America','United_State_Of_America': 'United State Of America', 'Us': 'United State Of America', '1': 'United State Of America', 'Usa': 'United State Of America',}
    data[f'{column}'] = data[f'{column}'].replace(city)
    return data[f'{column}']
```

```
data1 = pd.read_csv('only_wood_customer_us_1.csv')
data1.Gender = gender(data1)
data1.FirstName = names(data1,'FirstName')
data1.LastName = names(data1, 'LastName')
data1.UserName = names(data1, 'UserName')
data1.Email = data1.Email.str.lower()
data1.City = city(data1,'City')
data1.Country = country(data1, 'Country')
data1.rename(columns={'LastName': 'Lastname'}, inplace=True)
```
